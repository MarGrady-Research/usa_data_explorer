```python
# Import packages
import pandas as pd
import os
from dotenv import load_dotenv
from sqlalchemy import create_engine
```

```python
# Import data
analysis = pd.read_parquet("../data/test_percentiles.parquet")

# load secret variables from .env file
load_dotenv()

# grab URL
DB_URI = os.getenv("SUPABASE_DB_URL")
```

```python
# Upload to Supabase PosgreSQL databse

# create URI (IF THIS CODE WILL BE PUBLIC, WHERE DO I STORE THE PASSWORD?)
DB_URI = os.getenv("SUPABASE_DB_URL")

# Create the connection engine
engine = create_engine(DB_URI)

# Upload  dataframe
print("Uploading data to the cloud... please wait.")
analysis.to_sql(
    name='test_percentiles', 
    con=engine, 
    if_exists='replace', 
    index=False,
    chunksize=5000
)
print("Data successfully uploaded!")
```
